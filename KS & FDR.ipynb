```python
import pandas as pd
import numpy as np
import scipy.stats as sps
data=pd.read_csv('final data.csv')
```

```python
data=data.dropna()
```

```python
#去掉非P
for i in data.index:
    if data.at[i,'Transtype'] != 'P':
        data.drop(i)
```

```python
# add random column
data['Random'] = 0
for i in data.index:
    data.at[i,'Random'] = np.random.randint(100)
```

```python
# 分隔OTT，以后的内容用data_test
a=0
for i in data.index:
    if str(data.at[i,'Date']) == '2010-11-01 00:00:00':
        a = i
        break

data_test = pd.DataFrame(columns=data.columns)
k=0
for i in data.index:
    if i < a:
        k+=1
k
data_test = data.head(k)
```

```python
nvars = len(data.columns)
nvars
```

```python
goods = data[data['Fraud']==0]
bads = data[data['Fraud']==1]
numgoods = len(goods)
numbads = len(bads)
```

```python
KS = pd.DataFrame(np.zeros((nvars,3)))
i = 0
for col in data:
    KS.loc[i,0] = col
    i+=1
KS
```

```python
KS.columns = ['field','KS','FDR']
KS
```

```python
%%time
i = 0
for col in data:
    KS['KS'][i] = sps.ks_2samp(goods[col],bads[col])[0]
    i+=1
```

```python
%%time
topRows = int(round(len(data)*0.03))
j = 0
for col in data:
    temp1 = data.sort_values(col,ascending=False).head(topRows)
    temp2 = data.sort_values(col,ascending=True).head(topRows)
    needed1 = temp1.loc[:,'Fraud']
    needed2 = temp2.loc[:,'Fraud']
    FDR1 = sum(needed1)/numbads
    FDR2 = sum(needed2)/numbads
    FDRate = np.maximum(FDR1,FDR2)
    KS.loc[j,'FDR'] = FDRate
    j = j+1
```
